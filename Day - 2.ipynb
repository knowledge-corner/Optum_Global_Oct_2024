#### Bullet chart

```python
cols = np.where(profits >= target_profits, "palegreen", "coral")
```

```python
plt.figure(figsize=(6, 4))
plt.barh(products, profits, height=0.7, edgecolor = "black", linewidth = 0.4, color = cols)
plt.plot(target_profits, products, marker = "|", markersize = 15, linewidth = 0, color = "darkslategray")

plt.title("Profit Target Status across Products", size = 10, color = "darkslategray")
plt.grid(axis="x", color = "orange", ls = "--", linewidth = 0.2)

plt.yticks(size = 6,  color = "darkslategray")
y_ticks = np.arange(int(min(profits))-1000, max(profits) + 5000, 2000).astype(int)
y_labels = (y_ticks // 1000).astype(str) + "K"
plt.xticks(y_ticks, y_labels, size = 6,  color = "darkslategray")

# Annotating with %target achieved
percentage = (profits/target_profits * 100).astype(int)
percentage[percentage<0] = 0
for x, y, label in zip(profits, products, percentage):
    if label < 100 and x > 0 :
        plt.annotate(f"{label}%", xy = (x - 100, y), size = 6, color = "black", ha='right', va = "center")
    elif label < 100 and x < 0 :
        plt.annotate(f"{label}%", xy = (x + 200, y), size = 6, color = "black", ha='left', va = "center")
plt.show()
```
